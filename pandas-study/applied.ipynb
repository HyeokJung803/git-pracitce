```python
import pandas as pd

# 올바른 타이타닉 데이터셋 URL
TITANIC_URL = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# 데이터 불러오기
df = pd.read_csv(TITANIC_URL)

# 데이터의 처음 5행 확인
print(df.head())
```

```python
df.groupby('Pclass').agg(생존률 = ('Survived', 'mean'), 요금 = ('Fare', 'mean'), 인원 = ('Survived', 'count'))
```
